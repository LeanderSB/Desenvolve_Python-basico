### Q3.

Escreva um programa que lê 10 valores inteiros positivos do usuário e ao final imprime a média dos valores digitados com duas casas decimais.

```
Digite 10 números positivos:
3
2
12
14
20
42
19
0
8
1
A média dos valores digitados é 12.10
```

````python
## Escreva e execute seu código aqui
def calcular_media():

    soma = 0
    for i in range(10):
        while True:
            try:
                numero = int(input("Digite um número inteiro positivo: "))
                if numero > 0:
                    soma += numero
                    break
                else:
                    print("Número inválido. Digite um número inteiro positivo.")
            except ValueError:
                print("Entrada inválida. Digite um número inteiro.")

    media = soma / 10
    return f"{media:.2f}"

if __name__ == "__main__":
    resultado = calcular_media()
    print("A média dos números é:", resultado)
````

### Q5.

Faça um programa que lê dois inteiros N e M, e imprime na tela um campo de batalha naval. O tabuleiro deve possuir N linhas e M colunas. A primeira linha é composta por um espaço em branco e o cabeçalho das colunas, ou seja, valores de 1 a M. As N linhas seguintes iniciam com o cabeçalho da linha, ou seja, seu número, seguido de M caracteres "/" (barra) indicando uma possível posição jogável. 

```
Entrada:
5
4

Saída:
  1 2 3 4 
1 / / / / 
2 / / / / 
3 / / / / 
4 / / / / 
5 / / / / 
```

Para esse exercício, precisamos lembrar que o comando print implicitamente adiciona uma quebra de linha ao final da impressão. Podemos interferir no final da impressão adicionando mais uma entrada ao print. No exemplo, finalizamos cada linha com um espaço em branco:
```python
print("Texto qualquer", end = " ")
```

````python
## Escreva e execute seu código aqui

N = int(input("Digite o número de linhas (N): "))
M = int(input("Digite o número de colunas (M): "))


print(" ", end=" ")  
for col in range(1, M + 1):
    print(col, end=" ")
print() 


for row in range(1, N + 1):
    print(row, end=" ")  
    for col in range(1, M + 1):
        print("/", end=" ")  
    print()  
````

### Q6.

Escreva um programa que lê uma quantidade indefinida de valores e informa o maior e o menor valor digitados. A leitura é encerrada quando o usuário digitar o valor 0 (zero).

```
Entrada:
12
-1
29
11
-5
0

Saída:
Maior: 29
Menor: -5
```

````python
## Escreva e execute seu código aqui

maior = None
menor = None

print("Digite os valores (digite 0 para encerrar):")

while True:
    
    valor = int(input())

    
    if valor == 0:
        break

   
    if maior is None or valor > maior:
        maior = valor
    if menor is None or valor < menor:
        menor = valor


print(f"Maior: {maior}")
print(f"Menor: {menor}")
````

### Q7.

Escreva um programa que lê uma quantidade indefinida de valores e informa o **produto** dos valores positivos digitados. A leitura é encerrada quando o usuário digitar o valor 0 (zero).

```
Entrada:
2
-1
1
3
-5
0

Saída:
Produto: 6
```

````python
## Escreva e execute seu código aqui

produto = 1
tem_valor_positivo = False  

print("Digite os valores (digite 0 para encerrar):")

while True:
    
    valor = int(input())

   
    if valor == 0:
        break

    
    if valor > 0:
        produto *= valor
        tem_valor_positivo = True  


if tem_valor_positivo:
    print(f"Produto: {produto}")
else:
    print("Produto: 0")  
````

### Q8.

Vamos fazer uma calculadora que aceita expressões aritméticas de qualquer tamanho até que o usuário digite a palavra "Fim". Sua calculadora executa apenas as operações "+", "-".

```
Entrada:
5
+
8
-
3
+
2
Fim

Saída:
12
```

````python
## Escreva e execute seu código aqui
print("Digite a expressão, um elemento por vez (número ou operador). Digite 'Fim' para encerrar:")


resultado = 0
operador = "+"
primeira_entrada = True

while True:
    entrada = input()

    
    if entrada == "Fim":
        break

    
    if entrada in ["+", "-"]:
        
        operador = entrada
    else:
        
        numero = int(entrada)
        if primeira_entrada:  
            resultado = numero
            primeira_entrada = False
        else:
            if operador == "+":
                resultado += numero
            elif operador == "-":
                resultado -= numero


print(f"Resultado: {resultado}")
````
